```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.cm as cm

delta = 0.025
x = np.arange(-6.0, 6.0, delta)
y = np.arange(-6.0, 6.0, delta)
X, Y = np.meshgrid(x, y)
Z = (((X**2+Y-11)**2) + (((X+Y**2-7)**2)))
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z, 50)

ax.set_title('Himmelblau function')
```
